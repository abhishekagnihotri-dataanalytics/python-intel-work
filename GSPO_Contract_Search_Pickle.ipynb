```python
import pandas as pd
import re
contract = pd.read_pickle(r"C:\Users\jrstrong\Desktop\GSM\SOO\BI Projects\ContractMining\Contracts\df_contract_text.pkl")`
```

```python
searchPhrases = ['home sale referral rebate','service fee','calculation of rebate','rebate']
```

```python
def searchPhrase (example):
    termCnt = {}
    for term in searchPhrases:
        termCnt[term] = len([m.start() for m in re.finditer(r'\b' + term + r'\b', example)])
    return termCnt
```

```python
def unpack(df, column, fillna=None):
    import pandas as pd
    ret = None
    if fillna is None:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems()))], axis=1)
        del ret[column]
    else:
        ret = pd.concat([df, pd.DataFrame((d for idx, d in df[column].iteritems())).fillna(fillna)], axis=1)
        del ret[column]
    return ret
```

```python
contractResults = contract.to_frame()
contractResults = contractResults['Text'].apply(searchPhrase)
contractResults = contractResults.to_frame().reset_index()
contractResults.rename(columns = {'Text':'SearchTermCnt'},inplace=True)
contractResults = unpack(contractResults,'SearchTermCnt', 0)
contractResults['TotalCnt'] = 0
contractResults['TotalCnt'] = contractResults.drop('Document Id', 1).sum(axis=1).rename('TotalCnt')
contractResults = contractResults.loc[contractResults['TotalCnt'] > 0]
```
